```python
import random
import matplotlib.pyplot as plt

def Check(TF):
    door_L=[0,0,0]
    car=random.randint(0,2)
    door_L[car]=1
    k=0
    Ppu=random.randint(0,2)
    DH=random.randint(0,2)
    while((Ppu==DH)or(car==DH)):
        DH=random.randint(0,2)
    if TF=="T":
        NPpu=random.randint(0,2)
        while((NPpu==Ppu)or(NPpu==DH)):
            NPpu=random.randint(0,2)
        if NPpu==car:
            k+=1
    elif TF=="F":
        if Ppu==car:
            k+=1
    else:
        k=13
    return k

def MVeri(TF,n,w_l_Er):
    win=0
    lose=0
    Er=0

    for i in range(1,n+1):
        if Check(TF)==1:
            win+=1
        elif Check(TF)==0:
            lose+=1
        else:
            Er+=1
    
    if w_l_Er=="w":
        return win
    elif w_l_Er=="l":
        return lose
    else:
        return Er

r=int(input())
x=[]
y1=[]
y2=[]
for i in range(1,r+1):
    y2.append((MVeri("F",i,"w"))/i)
    y1.append((MVeri("T",i,"w"))/i)
    x.append(i)

fig=plt.figure(figsize=(6,4),facecolor="white")
plt.plot(x,y1,color="lightblue",label="True")
plt.plot(x,y2,color="red",label="False")
plt.legend(loc="upper right")
plt.show
```
